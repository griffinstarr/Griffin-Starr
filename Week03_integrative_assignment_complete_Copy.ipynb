## Part 1: Shipping Cost Calculator (25 points)

The most inconsistent calculation at SwiftShip is the basic shipping cost. Priya documents the official pricing rules:

**Base Rate by Zone:**
- Zone 1 (Local, < 100 miles): $5.99
- Zone 2 (Regional, 100-500 miles): $9.99
- Zone 3 (National, 500-1500 miles): $14.99
- Zone 4 (Extended, > 1500 miles): $24.99

**Weight Surcharge:**
- First 5 lbs: Included in base rate
- 5-20 lbs: $0.50 per lb over 5
- 20-50 lbs: $0.75 per lb over 20 (plus the 5-20 charge)
- Over 50 lbs: $1.00 per lb over 50 (plus previous charges)

**Service Level Multiplier:**
- Standard (5-7 days): 1.0x
- Express (2-3 days): 1.5x
- Overnight: 2.5x

**Your Task:**
Create three functions:
1. `get_base_rate(distance_miles)` - Returns the base rate for a given distance
2. `calculate_weight_surcharge(weight_lbs)` - Returns the weight surcharge
3. `calculate_shipping_cost(distance_miles, weight_lbs, service_level="standard")` - Returns the total shipping cost

The third function should call the first two.

```python
# Part 1: Shipping Cost Calculator
# Build the core pricing functions


def get_base_rate(distance_miles):
    if distance_miles < 100:
        return 5.99
    elif distance_miles < 500:
        return 9.99
    elif distance_miles <= 1500:
        return 14.99
    else:
        return 24.99

def calculate_weight_surcharge(weight_lbs):
    if weight_lbs <= 5:
        return 0.0
    elif weight_lbs <= 20:
        return (weight_lbs - 5) * 0.50
    elif weight_lbs <= 50:
        return (20 - 5) * 0.50 + (weight_lbs - 20) * 0.75
    else:
        return (20 - 5) * 0.50 + (50 - 20) * 0.75 + (weight_lbs - 50) * 1.00

def calculate_shipping_cost(distance_miles, weight_lbs, service_level="standard"):
    multipliers = {"standard": 1.0, "express": 1.5, "overnight": 2.5}
    base = get_base_rate(distance_miles)
    surcharge = calculate_weight_surcharge(weight_lbs)
    multiplier = multipliers.get(service_level.lower(), 1.0)
    return (base + surcharge) * multiplier
```

## Part 2: Delivery Time Estimator (20 points)

Customers always want to know: "When will my package arrive?" SwiftShip needs a consistent way to estimate delivery times.

**Business Days Calculation:**
- Standard: 5-7 business days → use 6 as estimate
- Express: 2-3 business days → use 2 as estimate  
- Overnight: 1 business day

**Additional Delays:**
- Distances over 1000 miles: Add 1 business day
- Packages over 50 lbs: Add 1 business day (requires special handling)
- Rural delivery (is_rural=True): Add 2 business days

**Your Task:**
Create two functions:

1. `estimate_delivery_days(distance_miles, weight_lbs, service_level, is_rural=False)` that:
   - Calculates the estimated business days for delivery
   - Returns both the minimum and maximum days as a tuple: `(min_days, max_days)`
     - For standard: range is estimate ± 1
     - For express: range is estimate to estimate + 1
     - For overnight: range is estimate to estimate

2. `calculate_delivery_date(min_days, max_days)` that:
   - Uses the `datetime` module to calculate actual delivery dates
   - Starts from today's date
   - **Skips weekends** (Saturday and Sunday are not business days)
   - Returns a tuple of `(earliest_date, latest_date)` as formatted strings (e.g., "February 15, 2026")

```python
# Part 2: Delivery Time Estimator

from datetime import datetime, timedelta

def estimate_delivery_days(distance_miles, weight_lbs, service_level, is_rural=False):
    base = {"standard": (5, 7), "express": (2, 3), "overnight": (1, 1)}
    min_days, max_days = base.get(service_level.lower(), (5, 7))
    if distance_miles > 1000:
        min_days += 1; max_days += 1
    if weight_lbs > 50:
        min_days += 1; max_days += 1
    if is_rural:
        min_days += 2; max_days += 2
    return (min_days, max_days)

def calculate_delivery_date(min_days, max_days):
    def add_business_days(start, n):
        current, counted = start, 0
        while counted < n:
            current += timedelta(days=1)
            if current.weekday() < 5:
                counted += 1
        return current

    today = datetime.now().date()
    earliest = add_business_days(today, min_days)
    latest = add_business_days(today, max_days)
    return (earliest.strftime("%B %d, %Y"), latest.strftime("%B %d, %Y"))
    # Your code here
    # Hint: Use datetime.now() to get today's date
    # Hint: Use timedelta(days=1) to add one day at a time
    # Hint: Use date.weekday() to check if a day is a weekend (5=Sat, 6=Sun)
    # Hint: Use strftime("%B %d, %Y") to format the date
    pass
```

## Part 3: Customer Quote Generator (25 points)

Now Priya needs to combine everything into a professional quote generator that the sales team can use.

**Quote Requirements:**
- Calculate shipping cost
- Calculate estimated delivery window
- Apply volume discount if quantity > 10 packages:
  - 11-25 packages: 5% discount
  - 26-50 packages: 10% discount
  - 51+ packages: 15% discount
- Return a dictionary containing all quote details

**Your Task:**
Create a function `generate_quote(distance, weight, service_level, quantity, is_rural=False)` that:
1. Uses your previous functions to calculate costs and delivery times
2. Applies volume discounts
3. Returns a dictionary with all quote information

```python
# Part 3: Customer Quote Generator

def calculate_volume_discount(quantity):
    if quantity > 50:
        return 0.15
    elif quantity > 25:
        return 0.10
    elif quantity > 10:
        return 0.05
    else:
        return 0.0

def generate_quote(distance, weight, service_level, quantity, is_rural=False):
    unit_cost = calculate_shipping_cost(distance, weight, service_level)
    subtotal = unit_cost * quantity
    discount_rate = calculate_volume_discount(quantity)
    discount_amount = subtotal * discount_rate
    total = subtotal - discount_amount
    min_days, max_days = estimate_delivery_days(distance, weight, service_level, is_rural)
    earliest, latest = calculate_delivery_date(min_days, max_days)
    return {
        "unit_cost": unit_cost,
        "quantity": quantity,
        "subtotal": subtotal,
        "discount_rate": discount_rate,
        "discount_amount": discount_amount,
        "total": total,
        "delivery_min": min_days,
        "delivery_max": max_days,
        "earliest_delivery": earliest,
        "latest_delivery": latest,
        "quote_date": datetime.now().strftime("%B %d, %Y"),
    }
    # Return a dictionary with:
    # - unit_cost: cost per package before discount
    # - quantity: number of packages
    # - subtotal: unit_cost * quantity
    # - discount_rate: volume discount rate
    # - discount_amount: dollar amount saved
    # - total: final cost after discount
    # - delivery_min: minimum delivery days
    # - delivery_max: maximum delivery days
    # - earliest_delivery: formatted earliest delivery date
    # - latest_delivery: formatted latest delivery date
    # - quote_date: today's date formatted
    pass
```

## Part 4: Batch Processing (15 points)

SwiftShip receives bulk orders from business customers. Priya needs a function to process multiple shipments at once and generate a summary report.

**Your Task:**
Create a function `process_batch(shipments)` that:
1. Takes a list of shipment dictionaries as input
2. Generates a quote for each shipment
3. Returns a summary dictionary containing:
   - Total number of packages across all shipments
   - Total cost before discounts
   - Total discounts applied
   - Grand total after discounts
   - List of all individual quotes

```python
# Part 4: Batch Processing

def process_batch(shipments):
    quotes = []
    total_packages, subtotal, total_discounts = 0, 0.0, 0.0
    for s in shipments:
        quote = generate_quote(s["distance"], s["weight"], s["service_level"],
                               s["quantity"], s.get("is_rural", False))
        quotes.append(quote)
        total_packages += s["quantity"]
        subtotal += quote["subtotal"]
        total_discounts += quote["discount_amount"]
    return {
        "total_packages": total_packages,
        "subtotal": subtotal,
        "total_discounts": total_discounts,
        "grand_total": subtotal - total_discounts,
        "quotes": quotes,
    }
    pass
```

```python
# Part 5: Documentation and Testing

def run_validation_tests():
    print("SWIFTSHIP PRICING SYSTEM - VALIDATION REPORT")
    print("=" * 60)
    all_passed = True

    def check(label, condition):
        nonlocal all_passed
        if not condition:
            all_passed = False
        print(f"  [{'PASS' if condition else 'FAIL'}] {label}")

    print("\nTest 1: Minimum distance (1 mile)")
    check("Zone 1 base rate = $5.99", get_base_rate(1) == 5.99)

    print("\nTest 2: Maximum weight (100 lbs)")
    check("100 lb surcharge = $80.00", abs(calculate_weight_surcharge(100) - 80.0) < 0.01)

    print("\nTest 3: Service level multipliers")
    std = calculate_shipping_cost(50, 3, "standard")
    check("express = 1.5x standard", abs(calculate_shipping_cost(50, 3, "express") - std * 1.5) < 0.01)
    check("overnight = 2.5x standard", abs(calculate_shipping_cost(50, 3, "overnight") - std * 2.5) < 0.01)

    print("\nTest 4: Rural delivery delay")
    min_u, max_u = estimate_delivery_days(200, 10, "standard", False)
    min_r, max_r = estimate_delivery_days(200, 10, "standard", True)
    check("Rural adds 2 days to min", min_r == min_u + 2)
    check("Rural adds 2 days to max", max_r == max_u + 2)

    print("\nTest 5: Volume discounts")
    check("No discount for 10",  calculate_volume_discount(10) == 0.0)
    check("5% discount for 11",  calculate_volume_discount(11) == 0.05)
    check("10% discount for 26", calculate_volume_discount(26) == 0.10)
    check("15% discount for 51", calculate_volume_discount(51) == 0.15)

    print("\n" + "=" * 60)
    print("ALL TESTS PASSED" if all_passed else "SOME TESTS FAILED")
    return all_passed


def print_function_documentation():
    print("\n" + "=" * 60)
    print("SWIFTSHIP PRICING LIBRARY - FUNCTION REFERENCE")
    print("=" * 60)
    docs = [
        ("get_base_rate(distance_miles)", "Returns zone-based base rate."),
        ("calculate_weight_surcharge(weight_lbs)", "Returns tiered weight surcharge."),
        ("calculate_shipping_cost(distance_miles, weight_lbs, service_level='standard')", "Returns total cost (base + surcharge) × multiplier."),
        ("estimate_delivery_days(distance_miles, weight_lbs, service_level, is_rural=False)", "Returns (min_days, max_days) business day window."),
        ("calculate_delivery_date(min_days, max_days)", "Returns (earliest, latest) as formatted date strings."),
        ("calculate_volume_discount(quantity)", "Returns discount rate as a decimal."),
        ("generate_quote(distance, weight, service_level, quantity, is_rural=False)", "Returns full quote dict with costs and delivery dates."),
        ("process_batch(shipments)", "Returns aggregate totals and list of individual quotes."),
    ]
    for name, purpose in docs:
        print(f"\n  {name}\n    {purpose}")
    print("\n" + "=" * 60)

    # Print documentation for each function
    # Your code here
    pass
```
