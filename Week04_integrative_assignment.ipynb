```python
# Import NumPy
import numpy as np

# Set random seed for reproducibility (so everyone gets the same data)
np.random.seed(4641)

# Generate simulated daily returns data
# 20 stocks, 252 trading days (1 year)
# Returns are in decimal form (e.g., 0.02 = 2% return)

stock_names = np.array([
    'APEX', 'BOLT', 'CYBER', 'DATA', 'EDGE',
    'FLUX', 'GRID', 'HYPER', 'INTEL', 'JADE',
    'KITE', 'LOGIC', 'MATRIX', 'NEXUS', 'ORBIT',
    'PULSE', 'QUANT', 'RELAY', 'SPARK', 'TITAN'
])

# Generate returns with different volatility profiles
# High-growth tech stocks have higher volatility
volatilities = np.array([0.02, 0.025, 0.03, 0.018, 0.022,
                         0.028, 0.015, 0.035, 0.02, 0.025,
                         0.032, 0.019, 0.024, 0.021, 0.027,
                         0.023, 0.031, 0.017, 0.026, 0.029])

# Expected annual returns (varying by stock)
expected_returns = np.array([0.12, 0.15, 0.18, 0.10, 0.14,
                             0.16, 0.08, 0.22, 0.11, 0.13,
                             0.20, 0.09, 0.15, 0.12, 0.17,
                             0.14, 0.19, 0.07, 0.16, 0.21])

# Generate 252 days of returns for each stock
# Shape: (252, 20) - rows are days, columns are stocks
daily_returns = np.random.normal(
    loc=expected_returns/252,  # Daily expected return
    scale=volatilities,        # Daily volatility
    size=(252, 20)
)

# Portfolio weights (how much is invested in each stock)
# These sum to 1.0 (100% of portfolio)
weights = np.array([0.08, 0.06, 0.07, 0.05, 0.04,
                    0.05, 0.04, 0.06, 0.05, 0.04,
                    0.05, 0.04, 0.06, 0.05, 0.05,
                    0.05, 0.06, 0.04, 0.03, 0.03])

print("Data loaded successfully!")
print(f"Daily returns shape: {daily_returns.shape}")
print(f"Number of stocks: {len(stock_names)}")
print(f"Trading days: {daily_returns.shape[0]}")
print(f"Portfolio weights sum: {weights.sum():.2f}")
```

## Part 3: Risk Metrics (25 points)

The investment committee is particularly interested in downside risk—how bad could things get?

**Your Task:**
Calculate the following risk metrics:

1. **Value at Risk (VaR) at 95% confidence**
   - What is the 5th percentile of daily returns?
   - Interpretation: "On 95% of days, the portfolio won't lose more than X%"

2. **Conditional VaR (Expected Shortfall)**
   - Average of returns that are worse than the VaR
   - Interpretation: "When losses exceed VaR, the average loss is X%"

3. **Maximum Drawdown**
   - The largest peak-to-trough decline in cumulative value
   - Hint: Calculate running maximum of cumulative returns, then find biggest drop

4. **Downside Deviation**
   - Standard deviation of only the negative returns
   - More focused measure of "bad" volatility

5. **Sortino Ratio**
   - Like Sharpe, but uses downside deviation instead of total volatility
   - (Annualized Return - Risk-Free Rate) / Annualized Downside Deviation

```python
# Part 3: Risk Metrics

# 1. Value at Risk (95%)
# Hint: Use np.percentile()


# 2. Conditional VaR (Expected Shortfall)
# Average of returns below VaR


# 3. Maximum Drawdown
# Calculate cumulative returns first, then the running maximum (track the largest
# cumulative value seen so far with a for-loop), then the drawdowns


# 4. Downside Deviation
# Standard deviation of only negative returns


# 5. Sortino Ratio
# 1. Value at Risk (95%)
var_95 = np.percentile(portfolio_daily_returns, 5)

# 2. Conditional VaR (Expected Shortfall)
cvar_95 = portfolio_daily_returns[portfolio_daily_returns <= var_95].mean()

# 3. Maximum Drawdown
running_max = np.zeros(252)
peak = cumulative_returns[0]
for i, val in enumerate(cumulative_returns):
    if val > peak:
        peak = val
    running_max[i] = peak

drawdowns = (cumulative_returns - running_max) / running_max
max_drawdown = drawdowns.min()

# 4. Downside Deviation
negative_returns = portfolio_daily_returns[portfolio_daily_returns < 0]
downside_dev = negative_returns.std() * np.sqrt(252)

# 5. Sortino Ratio
sortino_ratio = (port_ann_return - risk_free_rate) / downside_dev

print("RISK ANALYSIS REPORT")
print("="*60)
print("\nDownside Risk Metrics:")
print("-"*60)
print(f"Value at Risk (95%):          {var_95:.2%}")
print(f"  → On 95% of days, daily loss won't exceed this amount")
print(f"\nExpected Shortfall:           {cvar_95:.2%}")
print(f"  → When VaR is breached, average loss is this amount")
print(f"\nMaximum Drawdown:             {max_drawdown:.2%}")
print(f"  → Largest peak-to-trough decline over the period")
print(f"\nDownside Deviation (annualized): {downside_dev:.2%}")
print(f"\nSortino Ratio:                {sortino_ratio:.3f}")
print(f"  → Risk-adjusted return using only downside volatility")

# Display risk report
print("RISK ANALYSIS REPORT")
print("="*60)
print("\nDownside Risk Metrics:")
print("-"*60)
print(f"Value at Risk (95%): ")
print(f"  → On 95% of days, daily loss won't exceed this amount")
print(f"\nExpected Shortfall: ")
print(f"  → When VaR is breached, average loss is this amount")
print(f"\nMaximum Drawdown: ")
print(f"  → Largest peak-to-trough decline over the period")
print(f"\nDownside Deviation (annualized): ")
print(f"\nSortino Ratio: ")
print(f"  → Risk-adjusted return using only downside volatility")
```

## Part 4: Weight Optimization Scenario (20 points)

Sarah asks Jordan to explore what would happen with different portfolio weights.

```python
# Part 4: Weight Optimization Scenario

# Create equal weights


# Calculate equal-weight portfolio returns


# Calculate metrics for both portfolios
equal_weights = np.full(20, 0.05)

eq_daily_returns = (daily_returns * equal_weights).sum(axis=1)

eq_ann_return = eq_daily_returns.mean() * 252
eq_ann_vol = eq_daily_returns.std() * np.sqrt(252)
eq_sharpe = (eq_ann_return - risk_free_rate) / eq_ann_vol

# Max drawdown for equal weight
eq_cum = np.zeros(252)
val = 1.0
for i, r in enumerate(eq_daily_returns):
    val *= (1 + r)
    eq_cum[i] = val

eq_peak = eq_cum[0]
eq_max_dd = 0.0
for v in eq_cum:
    if v > eq_peak:
        eq_peak = v
    dd = (v - eq_peak) / eq_peak
    if dd < eq_max_dd:
        eq_max_dd = dd

print("PORTFOLIO COMPARISON")
print("="*60)
print(f"{'Metric':<25} {'Current':>15} {'Equal Weight':>15}")
print("-"*60)
print(f"{'Annualized Return':<25} {port_ann_return:>14.2%} {eq_ann_return:>14.2%}")
print(f"{'Annualized Volatility':<25} {port_ann_vol:>14.2%} {eq_ann_vol:>14.2%}")
print(f"{'Sharpe Ratio':<25} {port_sharpe:>14.3f} {eq_sharpe:>14.3f}")
print(f"{'Max Drawdown':<25} {max_drawdown:>14.2%} {eq_max_dd:>14.2%}")

weight_diff = weights - equal_weights
print("\nWEIGHT DIFFERENCES (Current vs Equal)")
print("-"*60)
for i in range(len(stock_names)):
    direction = "OVERWEIGHT" if weight_diff[i] > 0 else "UNDERWEIGHT"
    print(f"{stock_names[i]:<8} {weights[i]:.2%} vs {equal_weights[i]:.2%}  ({weight_diff[i]:+.2%}) {direction}")

print("\nINTERPRETATION")
print("-"*60)
if port_sharpe > eq_sharpe:
    print("The current weighted portfolio has a higher Sharpe ratio, meaning it delivers")
    print("better risk-adjusted returns than equal weighting. The overweights in higher-")
    print("quality stocks appear to be justified by the performance data.")
else:
    print("The equal-weight portfolio delivers a comparable or better Sharpe ratio,")
    print("suggesting the current active tilts may not be adding sufficient risk-adjusted")
    print("value to justify the concentration risk.")

# Compare portfolios
print("PORTFOLIO COMPARISON")
print("="*60)
print(f"{'Metric':<25} {'Current':>15} {'Equal Weight':>15}")
print("-"*60)
# Print comparison table


# Identify over/underweight positions
print("\nWEIGHT DIFFERENCES (Current vs Equal)")
print("-"*60)
# Your code here


# Interpretation
print("\nINTERPRETATION")
print("-"*60)
# Print your interpretation
```
